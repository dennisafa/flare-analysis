```python
from lightkurve import KeplerTargetPixelFile
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd

tpf = KeplerTargetPixelFile.from_archive('248432941')
```

```python
aper = np.zeros(tpf.shape[1:])
aper[16:19,11:17] = 1
tpf.plot(aperture_mask=aper.astype(bool), frame=200)
```

```python
lc = tpf.to_lightcurve(aperture_mask=aper.astype(bool))
pl.clf()

lc.flux = pd.rolling_median(lc.flux, 100, center=True)
pl.plot(lc.flux)
# pl.ylim(0.9991, 1.000
```

```python
for j in range(0, 18, 4):

    for i in range(0, 18, 4):
        aper = np.zeros(tpf.shape[1:])
        aper[i:i+4, j:j+4] = 1
        tpf.plot(aperture_mask=aper)
        
        pl.show()
        lc = tpf.to_lightcurve(aperture_mask=aper.astype(bool))
        lc.correct(windows=20, bins=10).bin(20).plot()
        pl.show()
        pl.clf()
```
